```python
pid = 1
label_map = {'AD':1, 'Healthy':0}
for folder in os.listdir(root):
    if folder == 'Readme.txt':
        break
    folder_path = os.path.join(root,folder)
    for filename in os.listdir(folder_path):
        # print(filename)
        labels[pid-1,0] = label_map[folder]
        labels[pid-1,1] = pid
        pid += 1
    # print(folder_path)
```

#### Process the data

```python
feature_path = 'Processed/ADSZ/Feature'
if not os.path.exists(feature_path):
    os.makedirs(feature_path)
```

```python
pid = 1
label_map = {'AD':1, 'Healthy':0}
subseq_length = 128
stride = 64  # Half of the subsequence length for half-overlapping
root = 'ADSZ'

for folder in os.listdir(root):
    if folder == 'Readme.txt':
        break
    folder_path = os.path.join(root,folder)
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
        data = []
        with open(file_path, 'r') as file:
            for line in file:
                values = [float(x) for x in line.split()]
                data.append(values)
        data_array = np.array(data)

        # Calculate the number of subsequences that can be extracted
        num_subsequences = (data_array.shape[0] - subseq_length) // stride + 1
        # Extract the subsequences
        subsequences = [data_array[i * stride : i * stride + subseq_length, :] for i in range(num_subsequences)]
        features = np.array(subsequences)

        print(f"Filename: {filename}")
        print(f"Patient ID: {pid}")
        print("Raw data:", data_array.shape)
        print("Segmented data", features.shape)
        np.save(feature_path + '/feature_{:02d}.npy'.format(pid), features)
        print('--------------------------------------------------------------------------')
        pid += 1
```
