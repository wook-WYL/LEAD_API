```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

#### Non-AD Datasets Ablation Study

```python
# Ensure the output directory exists
output_dir = "plots/"
os.makedirs(output_dir, exist_ok=True)  # Create directory if it doesn't exist


# Example custom x-axis labels
x_labels = [5, 7, 9, 11]
time = np.arange(len(x_labels))  # Convert custom labels into index positions (0, 1, 2, 3)

# Data for five subplots (each subplot has two lines)
data = [
    {"title": "ADFTD", 
     "line1": {"mean": [84.26, 82.81, 81.36, 79.96], "std": [2.90, 3.55, 3.55, 5.36], "label": "Subject-Level", "color": "red"},
     "line2": {"mean": [77.99, 76.77, 76.81, 76.64], "std": [1.14, 0.60, 0.60, 0.86], "label": "Sample-Level", "color": "blue"}},
    {"title": "BrainLat", 
     "line1": {"mean": [84.26, 84.26, 84.26, 89.98], "std": [2.90, 2.90, 2.90, 3.48], "label": "Subject-Level", "color": "red"},
     "line2": {"mean": [74.85, 74.81, 74.63, 77.80], "std": [0.44, 1.52, 1.17, 1.34], "label": "Sample-Level", "color": "blue"}},
    {"title": "CNBPM", 
     "line1": {"mean": [93.84, 96.92, 99.23, 100.00], "std": [3.08, 1.54, 1.54, 0.00], "label": "Subject-Level", "color": "red"},
     "line2": {"mean": [92.55, 94.12, 94.99, 95.53], "std": [1.08, 0.54, 0.40, 0.42], "label": "Sample-Level", "color": "blue"}},
    {"title": "Cognision-ERP", 
     "line1": {"mean": [73.29, 73.32, 71.62, 84.42], "std": [2.20, 2.84, 2.03, 2.21], "label": "Subject-Level", "color": "red"},
     "line2": {"mean": [67.20, 67.69, 67.46, 69.53], "std": [0.72, 0.75, 0.73, 0.91], "label": "Sample-Level", "color": "blue"}},
    {"title": "Cognision-rsEEG", 
     "line1": {"mean": [83.52, 86.32, 85.23, 91.86], "std": [0.08, 1.78, 1.35, 1.73], "label": "Subject-Level", "color": "red"},
     "line2": {"mean": [73.38, 74.28, 75.08, 76.01], "std": [0.75, 0.22, 0.28, 0.39], "label": "Sample-Level", "color": "blue"}}
]

# Initialize subplots: one row with five columns
fig, axes = plt.subplots(1, 5, figsize=(20, 4), sharey=True)

# Loop through each subplot and add lines
for i, ax in enumerate(axes):
    task_data = data[i]
    ax.set_title(task_data["title"])
    if i == 0:  # Add y-label only for the first subplot
        ax.set_ylabel("F1 Score %")
    ax.grid(True, linestyle="--", alpha=0.6)

    # Set y-axis limits and ticks
    ax.set_ylim(45, 105)
    ax.set_yticks(range(50, 101, 10))  # Set ticks from 50 to 100 with a step of 10

    # Plot the first line
    line1 = task_data["line1"]
    ax.plot(time, line1["mean"], label=line1["label"], color=line1["color"], marker="o")
    ax.fill_between(time,
                    np.array(line1["mean"]) - np.array(line1["std"]),
                    np.array(line1["mean"]) + np.array(line1["std"]),
                    color=line1["color"], alpha=0.2)

    # Plot the second line
    line2 = task_data["line2"]
    ax.plot(time, line2["mean"], label=line2["label"], color=line2["color"], marker="o")
    ax.fill_between(time,
                    np.array(line2["mean"]) - np.array(line2["std"]),
                    np.array(line2["mean"]) + np.array(line2["std"]),
                    color=line2["color"], alpha=0.2)

    # Set custom x-axis labels
    ax.set_xticks(time)
    ax.set_xticklabels(x_labels, fontsize=8)
    ax.set_xlabel("Number of Datasets", fontsize=8)

    # Add legend for each subplot
    ax.legend(loc="lower right", fontsize=8)

# Adjust layout to prevent overlapping
plt.tight_layout()
plt.subplots_adjust(left=0.15, right=0.98, top=0.9, bottom=0.3)

# Save the plot with tight bounding box
output_path = os.path.join(output_dir, "Non-AD-Ablation.png")
plt.savefig(output_path, dpi=300, bbox_inches="tight")
plt.show()

print(f"Plot saved to {output_path}")
```

#### AD Datasets Ablation Study

```python

# Ensure the output directory exists
output_dir = "plots/"
os.makedirs(output_dir, exist_ok=True)  # Create directory if it doesn't exist

# Data for ADFTD subplot
x_labels = [1, 2, 3, 4, 5]
time = np.arange(len(x_labels))  # Convert custom labels into index positions (0, 1, 2, 3, 4)

adftd_data = {
    "title": "ADFTD",
    "line1": {"mean": [82.81, 76.94, 86.25, 89.35, 86.31], "std": [3.55, 3.05, 3.15, 4.00, 3.27], "label": "Subject-Level", "color": "red"},
    "line2": {"mean": [73.75, 77.73, 80.84, 81.76, 82.39], "std": [1.00, 1.50, 0.93, 0.89, 0.73], "label": "Sample-Level", "color": "blue"}
}

# Initialize subplot
fig, ax = plt.subplots(figsize=(6, 4))

# Plot Subject-Level line
line1 = adftd_data["line1"]
ax.plot(time, line1["mean"], label=line1["label"], color=line1["color"], marker="o")
ax.fill_between(time,
                np.array(line1["mean"]) - np.array(line1["std"]),
                np.array(line1["mean"]) + np.array(line1["std"]),
                color=line1["color"], alpha=0.2)

# Plot Sample-Level line
line2 = adftd_data["line2"]
ax.plot(time, line2["mean"], label=line2["label"], color=line2["color"], marker="o")
ax.fill_between(time,
                np.array(line2["mean"]) - np.array(line2["std"]),
                np.array(line2["mean"]) + np.array(line2["std"]),
                color=line2["color"], alpha=0.2)

# Set title and labels
ax.set_title(adftd_data["title"], fontsize=10)
ax.set_ylabel("F1 Score %", fontsize=10)
ax.set_xlabel("Number of Datasets", fontsize=10)

# Set custom x-axis labels
ax.set_xticks(time)
ax.set_xticklabels(x_labels, fontsize=9)

# Set y-axis limits and ticks
ax.set_ylim(45, 105)
ax.set_yticks(range(50, 101, 10))

# Add legend
ax.legend(loc="lower right", fontsize=8)

# Grid and layout adjustment
ax.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()

# Save the plot
output_path = os.path.join(output_dir, "AD-Ablation.png")
plt.savefig(output_path, dpi=300, bbox_inches="tight")
plt.show()

print(f"Plot saved to {output_path}")
```
